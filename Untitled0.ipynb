# 3. Calculadora de Segundos

```python
dias = int(input('Digite a quantidade de dias: '))
horas = int(input('Digite a quantidade de horas: '))
minutos = int(input('Digite a quantidade de minutos: '))
segundos = int(input('Digite a quantidade de segundos: '))
dias
total_segundos = (dias * 24 * 60 * 60) + (horas * 60 * 60) + (minutos * 60) + segundos

print(f'O total de segundos é: {total_segundos}')
```

# 4. Calculo de Aumento Salarial

```python
salario_atual = float(input('Digite o salário atual: '))
percentual_aumento = float(input('Digite a porcentagem de aumento (ex: 10 para 10%): '))

valor_aumento = salario_atual * (percentual_aumento / 100)
novo_salario = salario_atual + valor_aumento

print(f'Valor do aumento: R$ {valor_aumento:.2f}')
print(f'Novo salário: R$ {novo_salario:.2f}')
```

# 5. Calculadora de Desconto

```python
preco_mercadoria = float(input('Digite o preço da mercadoria: '))
percentual_desconto = float(input('Digite o percentual de desconto (ex: 10 para 10%): '))

valor_desconto = preco_mercadoria * (percentual_desconto / 100)
preco_final = preco_mercadoria - valor_desconto

print(f'Valor do desconto: R$ {valor_desconto:.2f}')
print(f'Preço final a pagar: R$ {preco_final:.2f}')
```

# 6. Tempo de Viagem

```python
distancia_km = float(input('Digite a distância da viagem em km: '))
velocidade_media_kmh = float(input('Digite a velocidade média esperada em km/h: '))

tempo_total_horas = distancia_km / velocidade_media_kmh

horas_inteiras = int(tempo_total_horas)
minutos = round((tempo_total_horas - horas_inteiras) * 60)

print(f'O tempo estimado de viagem é de {horas_inteiras} horas e {minutos} minutos.')
```

# 8. Redução de Tempo de Vida de um Fumante.

```python
cigarros_por_dia = int(input('Quantidade de cigarros fumados por dia: '))
anos_fumando = int(input('Quantidade de anos que o fumante já fumou: '))

# Calculando o total de cigarros fumados
cigarros_total = cigarros_por_dia * 365 * anos_fumando

# Calculando os minutos de vida perdidos (10 minutos por cigarro)
minutos_perdidos = cigarros_total * 10

# Convertendo minutos perdidos para dias
dias_perdidos = minutos_perdidos / (24 * 60)

print(f'O fumante perderá aproximadamente {dias_perdidos:.2f} dias de vida.')
```
